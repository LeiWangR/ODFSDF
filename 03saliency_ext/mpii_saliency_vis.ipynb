```python
sample = 'saliency_mpii_coomking_sample/s08-d02-cam-002_00000_056'
```

## Section 1: video frame images

```python
%matplotlib inline
import matplotlib.pyplot as plt
import glob
import cv2

plt.figure(figsize=(15,80))
filelist = glob.glob(sample + '/images/*.jpg')
# print(filelist)
for i, file in zip(range(len(filelist)), sorted(filelist)):
    # print(file)
    plt.subplot(46, 4, i+1)
    im = cv2.imread(file)
    plt.imshow(im)
    plt.xticks([])
    plt.yticks([])
    plt.title('Frame No. {}'.format(i+1))
```

## Section 2: predicted saliency maps

```python
plt.figure(figsize=(15,80))
filelist = glob.glob(sample + '/saliency/*.jpg')
# print(filelist)
for i, file in zip(range(len(filelist)), sorted(filelist)):
    # print(file)
    plt.subplot(46, 4, i+1)
    im = cv2.imread(file)
    plt.imshow(im)
    plt.xticks([])
    plt.yticks([])
    plt.title('Frame No. {}'.format(i+1))
```
