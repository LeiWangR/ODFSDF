```python
sample = 'ACL/video_data/Riding-Horse-009'
```

## Section 1: video frame images

```python
%matplotlib inline
import matplotlib.pyplot as plt
import glob
import cv2

plt.figure(figsize=(15,30))
filelist = glob.glob(sample + '/images/*.png')
# print(filelist)
for i, file in zip(range(len(filelist)), sorted(filelist)):
    # print(file)
    plt.subplot(15, 4, i+1)
    im = cv2.imread(file)
    plt.imshow(im)
    plt.xticks([])
    plt.yticks([])
    plt.title('Frame No. {}'.format(i+1))
```

## Section 2: ground truth saliency maps

```python
plt.figure(figsize=(15,30))
filelist = glob.glob(sample + '/maps/*.png')
# print(filelist)
for i, file in zip(range(len(filelist)), sorted(filelist)):
    # print(file)
    plt.subplot(15, 4, i+1)
    im = cv2.imread(file)
    plt.imshow(im)
    plt.xticks([])
    plt.yticks([])
    plt.title('Frame No. {}'.format(i+1))
```

## Section 3: predicted saliency maps

```python
plt.figure(figsize=(15,30))
filelist = glob.glob(sample + '/saliency/*.png')
# print(filelist)
for i, file in zip(range(len(filelist)), sorted(filelist)):
    # print(file)
    plt.subplot(15, 4, i+1)
    im = cv2.imread(file)
    plt.imshow(im)
    plt.xticks([])
    plt.yticks([])
    plt.title('Frame No. {}'.format(i+1))
```
