## Section 1: Sample visualization
To load the .mat files, using *scipy.io.loadmat()* function.

```python
import scipy.io as ioo

x = ioo.loadmat('annotation_sample/0122.mat')['I']

x.shape
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

imgplot = plt.imshow(x)
plt.xticks([])
plt.yticks([])
```

## Section 2: One video sample visualization

### 2.1 video frames

```python
%matplotlib inline
import matplotlib.pyplot as plt
import glob
import cv2

plt.figure(figsize=(15,30))
filelist = glob.glob('Diving-Side-005/images/*.png')
# print(filelist)
for i, file in zip(range(len(filelist)), sorted(filelist)):
    # print(file)
    plt.subplot(14, 4, i+1)
    im = cv2.imread(file)
    plt.imshow(im)
    plt.xticks([])
    plt.yticks([])
    plt.title('Frame No. {}'.format(i+1))
```

### 2.2 fixation (binary eye fixation maps)

```python
plt.figure(figsize=(15,30))
filelist = glob.glob('Diving-Side-005/fixation/maps/*.mat')
# filelist = glob.glob('Diving-Side-005/fixation/*.png')
# print(filelist)
for i, file in zip(range(len(filelist)), sorted(filelist)):
    # print(file)
    plt.subplot(14, 4, i+1)
    im = ioo.loadmat(file)['I']
    # im = cv2.imread(file)
    plt.imshow(im)
    plt.xticks([])
    plt.yticks([])
    plt.title('Frame No. {}'.format(i+1))
```

### 2.3 continuous saliency maps

```python
plt.figure(figsize=(15,30))
filelist = glob.glob('Diving-Side-005/maps/*.png')
# print(filelist)
for i, file in zip(range(len(filelist)), sorted(filelist)):
    # print(file)
    plt.subplot(14, 4, i+1)
    im = cv2.imread(file)
    plt.imshow(im)
    plt.xticks([])
    plt.yticks([])
    plt.title('Frame No. {}'.format(i+1))
```
